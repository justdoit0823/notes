```python
from mpl_toolkits.mplot3d import Axes3D

import matplotlib.pyplot as plt

from matplotlib import cm
import numpy as np
```

```python
fig = plt.figure(figsize=(10, 12))
ax = Axes3D(fig)

X = np.arange(-10, 10, 0.5)
Y = np.arange(-10, 10, 0.5)
X, Y = np.meshgrid(X, Y)
Z = X ** 2 + Y ** 2
DX = X * 2
DY = Y * 2

ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_zlim(0, 200)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)
plt.title("function z = x ** 2 + y ** 2")
plt.show()

fig1 = plt.figure(figsize=(8, 10), dpi=120)
q_ax = fig1.add_subplot(111)
q_ax.quiver(X, Y, DX, DY, units='width')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 12))
ax1 = Axes3D(fig)

X = np.arange(-10, 10, 0.5)
Y = np.arange(-10, 10, 0.5)
X, Y = np.meshgrid(X, Y)
Z = 3 * X + 2 * Y
DX = 3
DY = 2

ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_zlabel("z")

ax1.set_xlim(-10, 10)
ax1.set_ylim(-10, 10)
ax1.set_zlim(-50, 50)

surf = ax1.plot_surface(X, Y, Z, cmap=cm.coolwarm)
plt.title("function z = 2 * x + 3 * y")
plt.show()

fig1 = plt.figure(figsize=(8, 10), dpi=120)
q_ax1 = fig1.add_subplot(111)
q_ax1.quiver(X, Y, DX, DY, units='width')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 12))
ax2 = Axes3D(fig)

X = np.arange(-10, 10, 0.5)
Y = np.arange(-10, 10, 0.5)
X, Y = np.meshgrid(X, Y)
Z = np.exp(-X ** 2 - Y ** 2)
DX = np.exp(-X ** 2 - Y ** 2) * (-2 * X)
DY = np.exp(-Y ** 2 - Y ** 2) * (-2 * Y)

ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_zlabel("z")

ax2.set_xlim(-10, 10)
ax2.set_ylim(-10, 10)
ax2.set_zlim(0, 1)

surf = ax2.plot_surface(X, Y, Z, cmap=cm.coolwarm)
plt.title("function z = e(- x** 2 - y** 2)")
plt.show()

fig1 = plt.figure(figsize=(8, 10), dpi=120)
q_ax2 = fig1.add_subplot(111)
q_ax2.quiver(X, Y, DX, DY, units='width')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 12))
ax3 = Axes3D(fig)

X = np.arange(-10, 10, 0.5)
Y = np.arange(-10, 10, 0.5)
X, Y = np.meshgrid(X, Y)
Z = np.sin(3 * X + 2 * Y)
DX = 3 * np.cos(3 * X + 2 * Y) * X
DY = 3 * np.cos(3 * X + 2 * Y) * Y

ax3.set_xlabel("x")
ax3.set_ylabel("y")
ax3.set_zlabel("z")

ax3.set_xlim(-10, 10)
ax3.set_ylim(-10, 10)
ax3.set_zlim(-2, 2)

surf = ax3.plot_surface(X, Y, Z, cmap=cm.coolwarm)
plt.title("function z = sin(3 * x - 2 * y)")
plt.show()

fig1 = plt.figure(figsize=(8, 10), dpi=120)
q_ax3 = fig1.add_subplot(111)
q_ax3.quiver(X, Y, DX, DY, units='width')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

```python
fig = plt.figure(figsize=(10, 10))
ax = Axes3D(fig)

X = np.arange(-5, 5, 0.3)
Y = np.arange(-5, 5, 0.3)
X, Y = np.meshgrid(X, Y)
Z = X ** 2 + Y ** 2

ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_zlim(0, 60)

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)

plt.title("function z = x ** 2 + y ** 2")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
ax1 = Axes3D(fig)

X = np.arange(-5, 5, 0.3)
Y = np.arange(-5, 5, 0.3)
X, Y = np.meshgrid(X, Y)
Z = 3 * X + 2 * Y

ax1.set_xlim(-10, 10)
ax1.set_ylim(-10, 10)
ax1.set_zlim(-20, 30)

surf = ax1.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)

plt.title("function z = 2 * x + 3 * y")
fig.show()
```

```python
fig = plt.figure(figsize=(10, 10))
ax1 = Axes3D(fig)

X = np.arange(-5, 5, 0.3)
Y = np.arange(-5, 5, 0.3)
X, Y = np.meshgrid(X, Y)
Z = np.exp(-X ** 2 - Y ** 2)

ax1.set_xlim(-10, 10)
ax1.set_ylim(-10, 10)
ax1.set_zlim(0, 1)

surf = ax1.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)

plt.title("function z = e(- x** 2 - y** 2)")
fig.show()
```

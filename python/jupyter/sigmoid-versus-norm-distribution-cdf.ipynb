```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
from scipy.stats import norm
```

```python
x_range = (-10, 10)
sample_num = 200
x = np.linspace(x_range[0], x_range[1], num=sample_num)

ys = 1 / (1 + np.exp(-x))
yc = norm.cdf(x)
yc1 = norm.cdf(x, scale=1.5)
yc2 = norm.cdf(x, scale=1.65)
yc3 = norm.cdf(x, scale=2)
yc4 = norm.cdf(x, scale=3)
```

```python
plt.plot(x, ys, label='sigmoid')
plt.plot(x, yc, label='cdf-scale-1')
plt.plot(x, yc1, label='cdf-scale-1.5')
plt.plot(x, yc2, label='cdf-scale-1.65')
plt.plot(x, yc3, label='cdf-scale-2')
plt.plot(x, yc4, label='cdf-scale-3')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
plt.plot(stats.bernoulli.rvs(0.6, size=100))
```

```python
plt.plot(range(2), tuple(stats.bernoulli.pmf(i, 0.2) for i in range(2)), 'ob')
```

```python
plt.plot(range(2), tuple(stats.bernoulli.cdf(i, 0.2) for i in range(2)), 'ob')
```

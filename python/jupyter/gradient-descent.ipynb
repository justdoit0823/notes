```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def objective_func(x):
    return np.exp(np.sum(x ** 2))
```

```python
def jac(x):
    return 2 * objective_func(x) * x
```

```python
n_epoch = 100
r = [0.1, 0.05, 0.02, 0.01]
y = []

for i in range(len(r)):
    y0 = []
    x = np.random.randn(2) / 2
    for j in range(n_epoch):
        ret = objective_func(x)
        y0.append(ret)
        x = x - r[i] * jac(x)
    
    y.append(y0)
```

```python
df1 = pd.DataFrame(np.stack(y, axis=1), columns=r)
df1.plot()
```

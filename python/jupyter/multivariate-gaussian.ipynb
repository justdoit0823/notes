```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import norm, pearsonr
```

```python
sample_num = 1000
```

```python
n1 = norm(1, 0.5)
x1 = n1.rvs(sample_num)

n2 = norm(1, 1)
x2 = n2.rvs(sample_num)
```

```python
m1 = norm(2, 0.5)
y1 = m1.rvs(sample_num)

m2 = norm(2, 1)
y2 = m2.rvs(sample_num)
```

```python
# independent variates

df1_1 = pd.DataFrame(np.stack([x1, y1], axis=1), columns=['x', 'y'])
sns.jointplot('x', 'y', data=df1_1, kind='kde')
print('x1-y1 covariance', np.cov(x1, y1))
print('x1-y1 correlation', pearsonr(x1, y1))

df1_2 = pd.DataFrame(np.stack([x1, y2], axis=1), columns=['x', 'y'])
sns.jointplot('x', 'y', data=df1_2, kind='kde')
print('x1-y2 covariance', np.cov(x1, y2))
print('x1-y2 correlation', pearsonr(x1, y2))
```

```python
# independent variates

df2_1 = pd.DataFrame(np.stack([x2, y1], axis=1), columns=['x', 'y'])
sns.jointplot('x', 'y', data=df2_1, kind='kde')
print('x2-y1 covariance', np.cov(x2, y1))
print('x2-y1 correlation', pearsonr(x2, y1))

df2_2 = pd.DataFrame(np.stack([x2, y2], axis=1), columns=['x', 'y'])
sns.jointplot('x', 'y', data=df2_2, kind='kde')
print('x2-y2 covariance', np.cov(x2, y2))
print('x2-y2 correlation', pearsonr(x2, y2))
```

```python
# positive correlative variates

df3_1 = pd.DataFrame(np.stack([x1, x1 + y1], axis=1), columns=['x', 'y'])
sns.jointplot('x', 'y', data=df3_1, kind='kde')
print('x1-x1&y1 covariance', np.cov(x1, x1 + y1))
print('x1-x1&y1 correlation', pearsonr(x1, x1 + y1))

df3_2 = pd.DataFrame(np.stack([x1, x1 + y2], axis=1), columns=['x', 'y'])
sns.jointplot('x', 'y', data=df3_2, kind='kde')
print('x1-x1&y2 covariance', np.cov(x1, x1 + y2))
print('x1-x1&y2 correlation', pearsonr(x1, x1 + y2))
```

```python
# positive correlative variates

df4_1 = pd.DataFrame(np.stack([x2, x2 + y1], axis=1), columns=['x', 'y'])
sns.jointplot('x', 'y', data=df4_1, kind='kde')
print('x2-x2&y1 covariance', np.cov(x2, x2 + y1))
print('x2-x2&y1 correlation', pearsonr(x2, x2 + y1))

df4_2 = pd.DataFrame(np.stack([x2, x2 + y2], axis=1), columns=['x', 'y'])
sns.jointplot('x', 'y', data=df4_2, kind='kde')
print('x2-x2&y2 covariance', np.cov(x2, x2 + y2))
print('x2-x2&y2 correlation', pearsonr(x2, x2 + y2))
```

```python
# negative correlative variates

df5_1 = pd.DataFrame(np.stack([x1, np.exp(-x1) + y1 / 3], axis=1), columns=['x', 'y'])
sns.jointplot('x', 'y', data=df5_1, kind='kde')
print('x1-x1&y1 covariance', np.cov(x1, np.exp(-x1) + y1 / 3))
print('x1-x1&y1 correlation', pearsonr(x1, np.exp(-x1) + y1 / 3))

df5_2 = pd.DataFrame(np.stack([x1, np.exp(-x1) + y2 / 3], axis=1), columns=['x', 'y'])
sns.jointplot('x', 'y', data=df5_2, kind='kde')
print('x1-x1&y2 covariance', np.cov(x1, np.exp(-x1) + y2 / 3))
print('x1-x1&y2 correlation', pearsonr(x1, np.exp(-x1) + y2 / 3))
```

```python
# negative correlative variates

df6_1 = pd.DataFrame(np.stack([x2, np.exp(-x2) + y1 / 3], axis=1), columns=['x', 'y'])
sns.jointplot('x', 'y', data=df6_1, kind='kde')
print('x2-x2/y1 covariance', np.cov(x2, np.exp(-x2) + y1 / 3))
print('x2-x2/y1 correlation', pearsonr(x2, np.exp(-x2) + y1 / 3))

df6_2 = pd.DataFrame(np.stack([x2, np.exp(-x2) + y2 / 3], axis=1), columns=['x', 'y'])
sns.jointplot('x', 'y', data=df6_2, kind='kde')
print('x2-x2/y2 covariance', np.cov(x2, np.exp(-x2) + y2 / 3))
print('x2-x2/y2 correlation', pearsonr(x2, np.exp(-x2) + y2 / 3))
```
